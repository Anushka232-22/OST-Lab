```python
# PlantCare AI Analysis Code
# Display Results Only (No CSV Saving)

import random
import time
import pandas as pd

# Step 1: Simulate API responses
num_tests = 20

results = []

for i in range(num_tests):

    start_time = time.time()

    # Simulated confidence score
    confidence = round(random.uniform(0.1, 0.95), 2)

    # Simulated disease detection
    diseases = ["Leaf Spot", "Blight", "Healthy", "Rust"]
    disease = random.choice(diseases)

    time.sleep(random.uniform(0.2, 0.8))  # simulate delay

    end_time = time.time()
    response_time = round(end_time - start_time, 2)

    # Apply 30% confidence threshold
    if confidence >= 0.30:
        result_status = "Reliable Detection"
    else:
        result_status = "Uncertain Result"

    results.append({
        "Test": i+1,
        "Disease": disease,
        "Confidence": confidence,
        "Response Time (sec)": response_time,
        "Status": result_status
    })

# Step 2: Store results in DataFrame
df = pd.DataFrame(results)

# Step 3: Display Results
print("\nPlantCare AI Testing Results:\n")
print(df)

# Step 4: Summary Analysis
avg_time = df["Response Time (sec)"].mean()
reliable_count = df[df["Status"] == "Reliable Detection"].shape[0]
uncertain_count = df[df["Status"] == "Uncertain Result"].shape[0]

print("\n--- Analysis Summary ---")
print("Average Response Time:", round(avg_time, 2), "seconds")
print("Reliable Detections:", reliable_count)
print("Uncertain Results:", uncertain_count)
```
